## 1. Analyse de la Distribution

## 2. Relations entre Variables Numériques

## 4. Analyse des Outliers

- Détecter les valeurs extrêmes avec la méthode IQR
- Afficher les transactions concernées
- Représenter les outliers sur un graphique

```python
# Méthode IQR
Q1 = df['prix_vente'].quantile(0.25)
Q3 = df['prix_vente'].quantile(0.75)
IQR = Q3 - Q1

borne_inf = Q1 - 1.5 * IQR
borne_sup = Q3 + 1.5 * IQR

outliers = df[(df['prix_vente'] < borne_inf) | (df['prix_vente'] > borne_sup)]

print(outliers)

# Visualisation des outliers
plt.figure(figsize=(8,5))
plt.scatter(df.index, df['prix_vente'], label='Normal')
plt.scatter(outliers.index, outliers['prix_vente'], label='Outliers')
plt.xlabel('Index')
plt.ylabel('Prix de vente')
plt.title('Détection des outliers')
plt.legend()
plt.show()
```
