
### 2. Nettoyage des données
- Vérifiez s'il y a des valeurs manquantes dans le DataFrame.
- Si la colonne `prix_m2` contient des valeurs nulles, remplacez-les par la moyenne de la colonne.
- Supprimez les lignes qui ne possèdent pas de "Numéro de transaction".



```python


print(df[df.isna().any(axis=1)])

df["prix_vente"] = df["prix_vente"].fillna(df["prix_vente"].mean())

df = df[df['id_transaction'].notnull()]

df
```

### 3. Statistiques Centrales et de Dispersion
- Calculez la moyenne, la médiane et l'écart-type du prix de vente.
- Déterminez le prix minimum et le prix maximum enregistrés.
- Calculez le coefficient de variation (écart-type / moyenne) pour évaluer la dispersion des prix.

---


```python
print(f"Mean: {df["prix_vente"].mean()}")
print(f"Median: {df["prix_vente"].median()}")
print(f"Ecart-type: {df["prix_vente"].std()}")
print(f"Prix Min: {df["prix_vente"].min()}")
print(f"Prix Max: {df["prix_vente"].max()}")
```


### 4. Analyse par Segment (Agrégation)

```python
PrixMoyenParType = df.groupby("type_bien")["prix_vente"].mean()
print(PrixMoyenParType)
```
