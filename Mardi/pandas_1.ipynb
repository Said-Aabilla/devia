**Challenge 1** : Création et exploration d’un DataFrame

**Objectif** : Créer un DataFrame simple et explorer sa structure.

**Travail à faire** :

Crée un DataFrame avec 3 colonnes : Nom, Âge, Ville.

Liste des données: data=[\
                                                                   ["Amine", 28, "Casablanca"],
                                                                   ["Lina", 22, "Rabat"],
                                                                   ["Youssef", 35, "Fès"],
                                                                   ["Salma", 30, "Casablanca"],
                                                                   ["Nora", np.nan, "Tanger"]]
Affiche les 5 premières lignes avec head().
Affiche la structure du DataFrame avec info().

Affiche des statistiques descriptives avec describe().


```python
import pandas as pd
import numpy as np

data = [ ["Amine", 28, "Casablanca"],["Lina", 22, "Rabat"],["Youssef", 35, "Fès"],["Salma", 30, "Casablanca"], ["Nora", np.nan, "Tanger"] ]

df = pd.DataFrame(data, columns=["Nom", "Age", "Ville"])
df.head()
df.info()
df.describe()
```

**Challenge 2** : Sélection et filtrage de données
**Objectif** : Savoir sélectionner des colonnes/lignes et filtrer selon des conditions.

**Travail à faire **:

Sélectionne la colonne Ville du DataFrame.

Affiche toutes les lignes où l’Âge est supérieur à 25.

Affiche le nom et la ville des personnes de Casablanca.

```python
ageColumn = df[["Age"]]

ageColumnFiltered = df[df["Age"]>25]

print(ageColumnFiltered)

cityFiltered = df[df["Ville"] == "Casablanca" ][["Age", "Ville"]]
print(cityFiltered)



```

**Challenge 3** : Ajout et modification de colonnes

**Objectif **: Manipuler les colonnes du DataFrame.

**Travail à faire** :

Ajoute une colonne Année de Naissance = 2025 - Âge.

Modifie les noms en majuscules avec .str.upper().

Renomme la colonne Ville en Localisation.

```python


df["Année de Naissance"] = 2025 - df["Age"]

df["Nom"] = df["Nom"].str.upper()

df.rename(columns={'Ville': 'Localisation'})
```

**Challenge 4** : Gestion des valeurs manquantes
**Objectif** : Identifier et traiter les valeurs manquantes.

**Travail à faire** :

Introduis manuellement une valeur NaN dans la colonne Âge.

Affiche les lignes contenant des valeurs manquantes avec isnull().

Remplace les valeurs manquantes par la moyenne des âges avec fillna().

```python
print(df.isnull())

df["Age"] = df["Age"].fillna(df["Age"].mean())

df["Année de Naissance"] = 2025 - df["Age"]

print(df)
```

**Challenge 5** : Tri et suppression
**Objectif** : Trier un DataFrame et supprimer des lignes ou colonnes.

**Travail à faire :**

Trie les lignes par âge décroissant avec sort_values(by='Âge', ascending=False).

Supprime la colonne Année de Naissance avec drop().

Supprime la première ligne du DataFrame avec drop(index=0).

```python
df = df.sort_values(by="Age", ascending=False)
```
