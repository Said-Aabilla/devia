# Challenge : Nettoyage de Données

## Objectif
Remplacer les valeurs manquantes dans un DataFrame.

## Travail à faire
- Charger un DataFrame avec des valeurs manquantes
- Remplacer les NaN par la moyenne des colonnes
- Afficher le DataFrame nettoyé

```python
import pandas as pd
import numpy as np

data = {
    "A": [1, 2, np.nan, 4],
    "B": [5, np.nan, 7, 8],
    "C": [9, 10, 11, np.nan]
}

df = pd.DataFrame(data)

df = df.fillna(df.mean())

print(df)
```

# Challenge : Filtrage de Données

## Objectif
Filtrer les transactions > 1000.

```python
data = {
    "transaction_id": [1, 2, 3, 4],
    "montant": [500, 1500, 2000, 800]
}

df = pd.DataFrame(data)

df_filtre = df[df["montant"] > 1000]

print(df_filtre)
```

# Challenge : Agrégation de Données

## Objectif
Calculer les ventes par région.

```python
data = {
    "region": ["Nord", "Sud", "Nord", "Sud"],
    "ventes": [100, 200, 150, 300]
}

df = pd.DataFrame(data)

resultat = df.groupby("region")["ventes"].sum()

print(resultat)
```

# Challenge : Fusion de Données

## Objectif
Fusionner deux DataFrames sur customer_id.

```python
clients = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "nom": ["Ali", "Sara", "Youssef"]
})

commandes = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "montant": [100, 200, 300]
})

df = pd.merge(clients, commandes, on="customer_id")

print(df)
```

# Challenge : Tableaux Croisés Dynamiques

## Objectif
Créer un pivot table des ventes.

```python
data = {
    "produit": ["A", "A", "B", "B"],
    "region": ["Nord", "Sud", "Nord", "Sud"],
    "ventes": [100, 150, 200, 250]
}

df = pd.DataFrame(data)

pivot = pd.pivot_table(df, values="ventes", index="produit", columns="region", aggfunc="sum")

print(pivot)
```
